```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Baca dataset
df = pd.read_csv('/content/drive/MyDrive/Dataset/gender_submission.csv')
```

```python
# Informasi dasar tentang dataset: tipe data, jumlah data non-null, dll.
print(df.info())
```

```python
# Statistik deskriptif untuk setiap kolom numerik
print(df.describe())
```

```python
# Visualisasi distribusi fitur numerik
numerical_features = df.select_dtypes(include=['float64', 'int64']).columns.tolist()
```

```python
for feature in numerical_features:
    plt.figure(figsize=(12, 6))
    sns.histplot(df[feature], bins=30, kde=True)
    plt.title(f'Distribusi untuk {feature}')
    plt.show()
```

```python
# Visualisasi hubungan antara dua fitur menggunakan scatter plot untuk fitur-fitur yang relevan
sns.scatterplot(x='PassengerId', y='Survived', data=df)
plt.show()
```

```python
# Boxplot untuk fitur terhadap target/kategori tertentu, misal 'genre'
sns.boxplot(x='PassengerId', y='Survived', data=df)
plt.xticks(rotation=45)  # Rotasi label x-axis untuk keterbacaan yang lebih baik jika perlu
plt.show()
```

```python
# Korelasi antar fitur
corr_matrix = df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Korelasi Antar Fitur')
plt.show()
```
