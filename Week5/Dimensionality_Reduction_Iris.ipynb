```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Melihat dataset iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
# Membagi data menjadi training set dan test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Melakukan reduksi dimensi dengan PCA. Misalnya kita ingin mengurangi menjadi 2 komponen.
pca = PCA(n_components=2)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
# Melatih model Decision Tree menggunakan data yang sudah direduksi dimensinya
clf = DecisionTreeClassifier()
clf.fit(X_train_pca, y_train)
```

```python
# Melakukan prediksi test set
y_pred = clf.predict(X_test_pca)
```

```python
# Menghitung akurasi
accuracy = accuracy_score(y_test, y_pred)
print(f"Akurasi model: {accuracy * 100:.2f}%")
```

```python
# melakukan visualisasi data
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_train, cmap='viridis', edgecolor='k', s=150)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('2 Component PCA')
plt.show()
```
